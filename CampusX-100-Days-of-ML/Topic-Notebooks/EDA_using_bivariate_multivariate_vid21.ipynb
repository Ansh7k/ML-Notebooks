```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (6,4)
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

# Scatterplot (numerical-numerical)

```python
sns.scatterplot(data= tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex',style='smoker',size='size')
```

# barplot (numerical-categorical)

```python
titanic.head()
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

# boxplot(numerical-categorical)

```python
sns.boxplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

# distplot (numerical-categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# heatmap(categorical-categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

# clustermap(categorical-categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']), figsize=(6,4))
```
